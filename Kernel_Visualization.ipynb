# Kernel Visualization
In this code, I will perform the following steps: </br>
1- Generate some sample vibration data in two predefined classes.</br>
2- Run a CNN model over the generated data (the model is not optimized).</br>
3- Visualize the learnt kernels in the first layer of convolution. </br>

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import numpy as np
from scipy import interpolate, integrate
import scipy.signal as sig, hilbert
from scipy.fftpack import fft
from sklearn.preprocessing import OneHotEncoder
import random
from keras.models import Sequential
from keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from keras import optimizers
from keras.initializers import Ones, Constant
from keras.regularizers import l2
```

## Function to generate sample vibration data

```python
def signal_gen(time_tot, fs = 20000, oc = 3, fn = 150, decay_f = 2, n_samples = 200, speed_range = np.array([10,20])):
        
        # input info:
        #       time_tot: signal length in second
        #       fs: sampling rate
        #       oc: fault charecterisitc order
        #       fn: natual frequency
        #       decay_f: the decay parameter of fault impacts
        #       n_samples: total number of generated samples
        #       speed_range: range of speed
        
        
        
        # generate the time stamps 
        t = np.linspace(0, time_tot, num = int(fs*time_tot))
        
        
        
        # initialize the data: data.shape = number of samples * Len(t) + 2. the last column of data represents the data label 
        # (e.g. label = 1 for faulty signal and label = 0 for healthy signal). the one before the last column shows the speed data
        data = np.zeros([n_samples,len(t)+2])
        
        # lets generate samples one by one!
        for k in range(n_samples):
            
            # randomly chose the fault conidtion. label = 1: faulty signal and label = 0: healthy signal
            label = random.randint(0, 1) 
            
            # randomly select the speed
            fr = random.randint(speed_range[0], speed_range[1]) 
            
            
            if label == 1:     # faulty signal
            
                signal = np.zeros(len(t))    # initialize the signal value
                signal += np.sin(2*np.pi*fn*t) 
                
                signal += 0.4*np.random.normal(0,1,len(t))
                signal = signal / np.std(signal)   # normalize the signal
                
                data[k,0:len(t)] = signal
                data[k,len(t)] = fr    # add speed data
                data[k,len(t)+1] = label  # add label

            else:  # healthy signal 

                signal = np.zeros(len(t))    # initialize the signal value
                signal += np.sin(2*np.pi*(fn+4)*t) 
                
                signal += 0.4*np.random.normal(0,1,len(t))
                signal = signal / np.std(signal)     # normalize the signal
                
                data[k,0:len(t)] = signal
                data[k,len(t)] = fr  # add speed data
                data[k,len(t)+1] = label  # add label
                
                
            if 10 * k % n_samples == 0: 
                print ("Progress: %03d out of %03d samples have been generated" % (k, n_samples))
                
        return data
```

## Input Parameters

```python
fs = 1800    # sampling rate (usually much higher, but for simplicity I choose a low value)
fr = 5       # shaft speed in Hz. 
oc = 20      # fault charectersitic order
time_tot = 0.13        # signal time duration (usually longer than that, but for simplicity I chose low value)
fn = 150               # natural frequency of the system (usually much higher)
decay_f = 2            # decay factor  for gaussian decay model      
n_samples = 15000   # number of generated samples
```

```python
# generate data!
data = signal_gen(time_tot, fs, oc, fn, decay_f, n_samples, speed_range = np.array([fr,fr]))
print ('shape of data', data.shape)
```

## Lets plot two generated samples from two classes

```python
t = np.linspace(0, time_tot, num = int(fs*time_tot))
cont0 = True
cont1 = True
i = 0
while cont0 or cont1:
    if data[i,-1] == 0 and cont0 == True:
        plt.figure(figsize= [3,3], dpi = 100)
        plt.plot(t, data[i,:-2])
        plt.xlabel('time (s)')
        plt.grid(True)
        plt.ylabel('magnitude')
        cont0 = False
    if data[i,-1] == 1 and cont1 == True:
        plt.figure(figsize= [3,3], dpi = 100)
        plt.plot(t, data[i,:-2])
        plt.xlabel('time (s)')
        plt.grid(True)
        plt.ylabel('magnitude')
        cont1 = False
    i += 1
```

## Reshape data for model training

```python
trainX = np.expand_dims(data[:,:-2], axis = 2)  # the last two columns wont be used in input (obviously!)
trainy = pd.get_dummies(data[:,-1]).values
```

## Some model parameters
Many of them needs to be optimized

```python
verbose, epochs, batch_size = 1000, 50, 128
n_timesteps, n_features, n_outputs = trainX.shape[1], trainX.shape[2], trainy.shape[1]
first_kernel_size = 128
```

## Lets create a model in Keras
### You may get more beautiful results if you freeze the parameters in Dense layer... Model will focus on learning and relying on conv kernels.

```python
model = Sequential()
model.add(Conv1D(filters=4, kernel_size=first_kernel_size, activation='relu', input_shape=(n_timesteps,n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(rate = 0.3))

model.add(Conv1D(filters=2, kernel_size=16, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(rate = 0.3))

model.add(Flatten())
model.add(Dense(10, activation='relu', kernel_initializer = Constant(value=1/70), trainable=True, kernel_regularizer=l2(l=0.1)))
model.add(Dropout(rate = 0.3))

model.add(Dense(n_outputs, activation='softmax'))

# Use Adam if you prefer.
sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)

model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
model.summary()
```

## Train time!
### wont take more than 5 min on ml.c5.xlarge

```python
history = model.fit(trainX, trainy, epochs=epochs, batch_size=batch_size, verbose=verbose)
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])

plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.show()

# Plot training & validation loss values
plt.plot(np.log(history.history['loss']))
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

## Now, lets plot the kernels in the first layer

```python
filters, biases = model.layers[0].get_weights()  # I wont plot biasses here
print('shape of kernel:::', filters.shape)
```

###  Plot kernels in sample domain

```python

n_filters, ix = 4, 1

fig,a = plt.subplots(n_filters, ix, figsize=(4,7))
for i in range(n_filters):
    a[i].plot(filters[:, :, i], label = 'kernel'+str(i))
    ix += 1
    a[i].yaxis.grid(True)
    a[i].xaxis.grid(True)
    a[i].legend(loc='upper right')
a[i].set_xlabel('sample number')
plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)

plt.show()
```

### Plot kernels in time domain

```python

n_filters, ix = 4, 1
t = np.linspace(0,filters.shape[0]/fs,filters.shape[0])
fig,a = plt.subplots(n_filters, ix, figsize=(4,7))
for i in range(n_filters):
    a[i].plot(t, filters[:, :, i], label = 'kernel'+str(i))
    ix += 1
    a[i].yaxis.grid(True)
    a[i].xaxis.grid(True)
    a[i].legend(loc='upper right')
a[i].set_xlabel('time (s)')
plt.tight_layout(pad=1.08, h_pad=None, w_pad=None, rect=None)

plt.show()
```
